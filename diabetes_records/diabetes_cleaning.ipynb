```python
#identified that there are ? values for diag columns
df[['diag_1','diag_2','diag_3']] = df[['diag_1','diag_2','diag_3']].replace('?', np.nan)
```

```python
#function to turn values in diag columns to categories based on their codes
def categorize_icd9(code):
    try:
        if pd.isna(code):
            return 'Unknown'
        code = int(str(code).split('.')[0])
    except:
        return 'Unknown'
    # ICD-9 ranges mapped to categories
    if 1 <= code <= 139: return "Infectious diseases"
    elif 140 <= code <= 239: return "Neoplasms"
    elif 240 <= code <= 279: return "Endocrine/metabolic (incl. diabetes)"
    elif 280 <= code <= 289: return "Blood diseases"
    elif 290 <= code <= 319: return "Mental disorders"
    elif 320 <= code <= 389: return "Nervous system"
    elif 390 <= code <= 459: return "Circulatory system"
    elif 460 <= code <= 519: return "Respiratory system"
    elif 520 <= code <= 579: return "Digestive system"
    elif 580 <= code <= 629: return "Genitourinary system"
    elif 630 <= code <= 679: return "Pregnancy/childbirth"
    elif 680 <= code <= 709: return "Skin"
    elif 710 <= code <= 739: return "Musculoskeletal"
    elif 740 <= code <= 759: return "Congenital anomalies"
    elif 760 <= code <= 779: return "Perinatal conditions"
    elif 780 <= code <= 799: return "Symptoms/signs"
    elif 800 <= code <= 999: return "Injury/poisoning"
    else: return "Other/Unknown"
```

```python
#replaces the diag codes to the category
for col in ['diag_1','diag_2','diag_3']:
    df[col] = df[col].apply(categorize_icd9)
```

```python
#mapps the diag 1,2,3 intolong format
diagnoses_df = cleaned_df.melt(id_vars=['encounter_id'],
                               value_vars=['diag_1', 'diag_2', 'diag_3'],
                               var_name='diag_order',
                               value_name='icd9_code')
diagnoses_df['diag_order'] = diagnoses_df['diag_order'].map({
    'diag_1':1,
    'diag_2':2,
    'diag_3':3
})

```

```python
def map_diag_group(code):
    if pd.isnull(code):
        return 'Unknown'
    try:
        num = int(float(code))  # handles decimals like 250.13 → 250
    except ValueError:
        return 'Other'  # non-numeric codes like V45, E11

    if 390 <= num <= 459 or num == 785:
        return 'Circulatory'
    elif 460 <= num <= 519 or num == 786:
        return 'Respiratory'
    elif num == 250:
        return 'Diabetes'
    elif 520 <= num <= 579 or num == 787:
        return 'Digestive'
    elif 800 <= num <= 999:
        return 'Injury'
    elif 710 <= num <= 739:
        return 'Musculoskeletal'
    elif 140 <= num <= 239:
        return 'Neoplasms'
    else:
        return 'Other'

diagnoses_df['diag_group'] = diagnoses_df['icd9_code'].apply(map_diag_group)
```
