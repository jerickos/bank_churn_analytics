```python
df = df.drop(columns=['RowNumber', 'Surname'])
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Geography'] = df['Geography'].str.strip().str.title()
```

```python
df.head()
```

```python
#creating age groups
bins = [0, 25, 35, 45, 55, 75, 100] 
labels = ['18-24', '25-34', '35-44', '45-54', '55-74', '75+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
#cteaing credit score bands
bins_cs = [0, 500, 670, 740, 800, 900] 
labels_cs = ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent']
df['CreditScoreGroup'] = pd.cut(df['CreditScore'], bins=bins_cs, labels=labels_cs)
```

```python
df["HasBalance"] = df["Balance"].apply(lambda x: 1 if x > 0 else 0)
```
